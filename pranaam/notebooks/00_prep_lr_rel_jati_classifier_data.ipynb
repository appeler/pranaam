## Prepare LR Data For Name Classification

```python
import pandas as pd
```

```python
lr = pd.read_csv("../input/landrecords_deduped.csv", low_memory=False)
```

```python
lr.head()
```

```python
# father's last name
lr['last_fat'] = lr['name_of_father'].apply(lambda x: x if pd.isnull(x) else x.split(" ")[-1])
```

```python
lr.head()
```

## Join to caste codes + Produce unjoined castes

```python
# import caste codes
cc = pd.read_stata("../input/caste_codes/caste_code_landrecords.dta")
cc.head()
```

```python
print(cc.shape)
cc = cc.drop_duplicates(subset=['caste_landrecords'])
print(cc.shape)
```

```python
lr_cc = pd.merge(lr, cc, how = "left", left_on = "caste", right_on = "caste_landrecords")
print(lr.shape, lr_cc.shape)
```

```python
list(lr_cc)
```

```python
lr_cc['muslim'].value_counts()
```

```python
lr_cc_muslim = lr_cc.dropna(subset =['muslim'])
```

```python
lc_cc_muslim_subset = lr_cc_muslim[['name_of_ryot', 'name_of_father', 'caste_name', 'caste_category', 'muslim']]
lc_cc_muslim_subset_dedupe = lc_cc_muslim_subset.drop_duplicates()
```

```python
print(lc_cc_muslim_subset.shape)
lc_cc_muslim_subset_dedupe.shape
```

```python
# Long data + dedupe
lc_1 = lc_cc_muslim_subset_dedupe[['name_of_father', 'caste_name', 'caste_category', 'muslim']].rename(columns={"name_of_father": "name"})
lc_2 = lc_cc_muslim_subset_dedupe[['name_of_ryot', 'caste_name', 'caste_category', 'muslim']].rename(columns={"name_of_ryot": "name"})
lc_all = pd.concat([lc_1, lc_2]).dropna(subset =['name']).drop_duplicates()
```
